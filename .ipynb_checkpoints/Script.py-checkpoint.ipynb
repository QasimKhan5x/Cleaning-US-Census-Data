# Cleaning US Census Data

```python
import pandas as pd
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''Create the dataframe us_census'''
files = glob.glob('states[0-9].csv')
df_list = []
for file in files:
  df_list.append(pd.read_csv(file))
us_census = pd.concat(df_list)
```

## Cleaning the DataFrame

```python
us_census = us_census.drop(columns='Unnamed: 0')
```

```python
'''Cleaning Income Column'''

income = us_census.Income.str.replace('\$', '')
us_census['Income'] = pd.to_numeric(income)
```

```python
'''Remove GenderPop column by splitting it into Men and Women columns'''

genderpop = us_census['GenderPop'].str.split('_')
males = genderpop.str.get(0).str.replace('M', '')
females = genderpop.str.get(1).str.replace('F', '')
us_census['Men'] = pd.to_numeric(males)
us_census['Women'] = pd.to_numeric(females)
us_census = us_census.drop(columns='GenderPop')
```

```python
'''Fill nan values and drop duplicates'''

us_census = us_census.fillna({'Women': us_census.TotalPop - us_census.Men, 'Men': us_census.TotalPop - us_census.Women})
us_census = us_census.drop_duplicates()
```

```python
'''Clean ethnicity columns'''
ethnicities = list(us_census.columns)[2:8]
for ethnicity in ethnicities:
  us_census[ethnicity] = us_census[ethnicity].str.replace('%', '')
  us_census[ethnicity] = pd.to_numeric(us_census[ethnicity])
```

```python
'''Create a scatter plot of the data'''

fig, axes = plt.subplots(2, 1, figsize=(10, 8))
axes[0].scatter(us_census['Men'], us_census['Income'], color='r')
axes[0].set_ylabel('Income', fontsize=12, fontweight='semibold')
axes[0].set_xlabel('Male Population', fontsize=12, fontweight='semibold')
axes[1].scatter(us_census['Women'], us_census['Income'])
axes[1].set_ylabel('Income', fontsize=12, fontweight='semibold')
axes[1].set_xlabel('Female Population', fontsize=12, fontweight='semibold')
fig.suptitle('Income distribution by gender', fontsize=14, fontweight='semibold')
plt.show()
```
